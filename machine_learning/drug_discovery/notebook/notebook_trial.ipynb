```python
import os
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
```

```python
Draw.MolsToGridImage([Chem.MolFromSmiles(smi) for smi in df_combined['canonical_smiles'].iloc[0:9]])
```

```python
df_arr = Chem.SmilesMolSupplier('E:\VS Code\machine_learning\drug_discovery\data\processed\mmp9.smi',delimiter='\t',titleLine=False)
```

```python
fp = [ AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) for mol in df_arr if mol is not None]
```

```python
fp
```

```python
train_df = np.asarray(fp, dtype=np.int32)
```

```python
train_df
```

```python
ids = [mol.GetProp("_Name") for mol in df_arr if mol is not None]
```

```python
ids
```

```python
labels = np.asarray(ids, dtype=np.int32).reshape(-1,1)
```

```python
labels
```

```python
dataset = np.concatenate([train_df,labels],axis=1)
```

```python
np.save('E:\VS Code\machine_learning\drug_discovery\data\processed\mmp9_features',dataset)
```
